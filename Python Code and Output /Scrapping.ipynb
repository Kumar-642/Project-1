### Combining all stop words in a single List, let call *final_stop*

```python
all_stop = []
for i in range(1,8):
  stop = []
  filename = 'stop'+str(i)+'.txt'
  f_neg = open(filename,encoding = "ISO-8859-1")
  a = f_neg.readline()
  while(a!=''):
    stop.append(a[:-1])
    a = f_neg.readline()
  all_stop.append(stop)
```

```python
final_stop = []
for i in range(len(all_stop)):
  for j in range(len(all_stop[i])):
    if '|' in all_stop[i][j]:
      ind = all_stop[i][j].index('|')
      all_stop[i][j] = all_stop[i][j][:ind]
    final_stop.append(all_stop[i][j])         # final_stop contains all stop words.
```

### Extraction of Article from URL by web Scrapping

```python

# Extraction of Data From every URL

for i in range(len(URL_data)):
    r = requests.get(URL_data.URL[i])
    #r = requests.get('https://insights.blackcoffer.com/ai-in-healthcare-to-improve-patient-outcomes/')
    soup = bs(r.content)
    contents = soup.prettify()
    filename = str(URL_data.URL_ID[i])+'.txt'
    file = open(filename,'a')
    try:
      title_50 = soup.find(class_="td_block_wrap tdb_single_content tdi_130 td-pb-border-top td_block_template_1 td-post-content tagdiv-type")
      rank_50 = title_50.find_all('p')
    except:
      try:
        title_50 = soup.find(class_="td-post-content tagdiv-type")
        rank_50 = title_50.find_all('p')
      except:
        pass
    for row in rank_50 :
      n1 = row.text.strip()
      file.write(n1)
    file.close()
```

```python
# Total Number of words in every URL without deleting stop words

Total_Words = []
for i in range(37,151):
    TC = 0
    filename = str(i)+'.txt'
    file = open(filename,'r')
    for i in file:
      list_1 = i.split()
      TC+=len(list_1)
    Total_Words.append(TC)
```

```python
# Number of stop words in each URL article
Total_stop = []
for i in range(37,151):
    filename = str(i)+'.txt'
    file = open(filename,'r')
    for i in file:
      list_1 = i.split()
      for k in range(len(list_1)):
        if '.' in list_1[k]:
          list_1[k] = list_1[k][:-1]
    count_stop = 0
    non_count = 0
    for i in list_1:
      for j in final_stop:
        if i==j:
          count_stop+=1
    Total_stop.append(count_stop)
```

```python
# Negative List
negative_word = []
f_neg = open('negative-words.txt',encoding = "ISO-8859-1")
a = f_neg.readline()
while(a!=''):
  negative_word.append(a[:-1])
  a = f_neg.readline()
```

```python
# Positive List
positive_word = []
f_neg = open('positive-words.txt',encoding = "ISO-8859-1")
a = f_neg.readline()
while(a!=''):
  positive_word.append(a[:-1])
  a = f_neg.readline()
```

```python
# Finding negative score for every Link
NEGATIVE_SCORE = []
for i in range(37,151):
    sub_list = []
    filename = str(i)+'.txt'
    file = open(filename,'r')
    for i in file:
      list_1 = i.split()
      for k in range(len(list_1)):
        if '.' in list_1[k]:
          list_1[k] = list_1[k][:-1]
      sub_list+=list_1
    Neg_Word = 0
    for i in sub_list:
      for j in negative_word:
        if i==j and i not in final_stop:        # we do not consider those word which are in final-stop
          Neg_Word-=1
    NEGATIVE_SCORE.append(Neg_Word)
```

```python
NEGATIVE_SCORE
```

```python
# Finding Positive Score score for every Link
POSITIVE_SCORE = []
for i in range(37,151):
    subp_list = []
    filename = str(i)+'.txt'
    file = open(filename,'r')
    for i in file:
      list_1 = i.split()
      for i in range(len(list_1)):
        if '.' in list_1[i]:
          list_1[i] = list_1[i][:-1]
      subp_list+=list_1
    Pos_Word = 0
    for i in subp_list:
      for j in positive_word:   # we do not consider those word which are in final-stop
        if i==j and i not in final_stop:
          Pos_Word+=1
    POSITIVE_SCORE.append(Pos_Word)
```

```python
data = {'NEGATIVE SCORE' : NEGATIVE_SCORE,'POSITIVE SCORE' : POSITIVE_SCORE}
```

```python
# Calculating Polarity Score

Output['POLARITY SCORE'] = (Output['POSITIVE SCORE'] - Output['NEGATIVE SCORE'])/ ((Output['POSITIVE SCORE'] + Output['NEGATIVE SCORE']) + 0.000001)
```
